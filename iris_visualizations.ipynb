```python
# Importing libraries
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

##Importing Iris flower dataset in a pandas DataFrame
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
##Visulaizations using lirary "seaborn"

#Simple scatter_plot
iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
```

```python
#Seaborn jointplot: bivariate scatterplot + univariate histogram
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, size=6)
```

```python
#Scatter_plot: Fill color - Species
sns.FacetGrid(iris, hue="species", size=6) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
```

```python
#Boxplot: Compare the spread of individual features
ax = sns.boxplot(x="species", y="petal_length", data=iris)
#Adding point using jitter = True
ax = sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="red")
```

```python
#Violin_plot: Scatter Plot + Box Plot
sns.violinplot(x="species", y="petal_length", data=iris, size=6)
```

```python
## kde plot: To visualize a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="species", size=6) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()
```

```python
#pair plot, to check the bi-variate relationship of every pair of features
sns.pairplot(iris, hue="species", size=3)
```

Interpretation: From the pair plot we can see that species: setosa has most different features when compared to the other two                   species 

```python
# Pair_plot with diag elements as kde_plot
sns.pairplot(iris, hue="species", size=3, diag_kind="kde")
```

```python
##Visulaizations using Pandas
#Boxplot
iris.boxplot(by="species", figsize=(18, 12))
```

```python
#Andrews Curves: plotting the Fourier series generated using attributes of samples as coefficients
from pandas.plotting import andrews_curves
andrews_curves(iris, "species")
```

```python
# Radviz: Multivariate visualization technique, puts each feature as a 
# point on a 2D plane, and then simulates having each sample attached 
# to those points through a spring weighted by the relative value for 
# that feature
from pandas.plotting import radviz
radviz(iris, "species")
```
